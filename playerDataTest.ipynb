```python
from nba_api.stats.endpoints import leaguedashplayerstats

player_stats = leaguedashplayerstats.LeagueDashPlayerStats(per_mode_detailed="PerGame")
df = player_stats.get_data_frames()[0]

print(df.shape)
print(df.columns)
print(df.isnull().sum())
df.head()
```

```python
df = df[['PLAYER_ID', 'PLAYER_NAME', 'NICKNAME', 'TEAM_ABBREVIATION', 'GP', 'MIN', 'PTS', 'REB', 'AST', 'STL', 'BLK', 'FG_PCT', 'FT_PCT', 'FG3_PCT', 'FG3A']]
# df = df[(df['GP'] > 30)]
df = df.sort_values('FG3A', ascending=False)
print(df["TEAM_ABBREVIATION"].unique())
df.head()
```

```python
df.to_csv('player_data.csv', index=False)
```
